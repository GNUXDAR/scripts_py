### PROTECCIÓN DE ERRORES
Modifica el ejercicio anterior para evitar errores si el usuario escribe mal el nombre del archivo, cuando falle mostraremos un mensaje que diga, No se ha encontrado el archivo {nombrearchivo}

```
# TU CÓDIGO AQUÍ
```

<details>
<summary>Pincha aqui para ver la solución</summary>
<pre><code>
nombrearchivo = input('Seleccione que fichero quiere tratar:')
try:
  manejador = open('ejercicio.csv')
  cuenta = 0
  for linea in manejador:
    cuenta = cuenta+1
  print(f'el archivo {nombrearchivo} tiene {cuenta} lineas')
  manejador.close()
except:
  print(f'No se ha encontrado el archivo {nombrearchivo}')
