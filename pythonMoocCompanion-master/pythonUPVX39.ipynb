#### CONTAR LAS LINEAS DE UN ARCHIVO
Combina todo lo aprendido para contar las lineas que tiene el archivo ejercicio.csv

```python
# TU CÓDIGO AQUÍ
fichero = 'ejercicio.csv'
archivo = open(fichero, "r")
lineas = 0
for linea in archivo.readlines():
    lineas+=1
print(f'El archivo tiene {lineas} lineas')
archivo.close()
```

## SELECCIONANDO ARCHIVOS
Modifica el primer ejercicio para que nos permita seleccionar que archivo queremos utilizar al contar las lineas usando el comando input.

```python
# TU CÓDIGO AQUÍ
fichero = input('Ingrese el nombre del archivo que quiere contar las lineas: ')
archivo = open(fichero, "r")
lineas = 0
for linea in archivo.readlines():
    lineas+=1
print(f'El archivo {fichero} tiene {lineas} lineas')
archivo.close()
```

### PROTECCIÓN DE ERRORES
Modifica el ejercicio anterior para evitar errores si el usuario escribe mal el nombre del archivo, cuando falle mostraremos un mensaje que diga, No se ha encontrado el archivo {nombrearchivo}

```python
# TU CÓDIGO AQUÍ
try:
    fichero = input('Ingrese el nombre del archivo que quiere contar las lineas: ')
    archivo = open(fichero, "r")
    lineas = 0
    for linea in archivo.readlines():
        lineas+=1
    print(f'El archivo {fichero} tiene {lineas} lineas')
    archivo.close()
except:
    print(f'No se ha encontrado el archivo {fichero}')
```
