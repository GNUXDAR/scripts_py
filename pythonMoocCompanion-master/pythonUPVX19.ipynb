# Ejemplo de bloque if

```python
edad = 10
if edad > 18:
  print('puede beber en España')
print ('eso es todo')
```

```python
edad = 10
if edad > 18:
  print('puede beber en España')
  if edad > 67:
    print('y además estará jubilado')
print ('eso es todo')
```

```python
edad = 15
if edad > 18:
  print('puede beber en España')
else:
  print('aún no puede beber en España')
print('eso es todo')
```

Ejemplo con bloque for

```python
frutas = ['manzana', 'pera', 'platano']

print('comienza el bloque')
for fruta in frutas:
  print(fruta)
print('finaliza el bloque')
```
