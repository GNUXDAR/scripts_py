# Contextos

```
variableGlobal = 7
```

```
def contextoDefuncion(parametro):
  variable = 7
  print(f'la variable global que hemos definido antes valia {variableGlobal}')
  print(f'el valor de la variable local es {variable}')
  print(f'y el parámetro con el que hemos llamado a la función vale {parametro}')
```

```
contextoDefuncion('entrada')
```

```
print(f'la variable global que hemos definido antes valia {variableGlobal}')
```
